```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
## read in data
home = os.getcwd()
os.path.dirname(home)
df = pd.read_csv(f'{os.path.dirname(home)}/data/sleepstudy.csv')
df.head()
```

```python
# explore response variable
sns.histplot(df['Reaction'])
```

```python
# explore input variable
plt.figure(figsize=(10,6))
sns.lineplot(x='Reaction', y='Days', data = df)
```

```python
# explore repeated measures - do we have enough data per subject?
plt.figure(figsize=(10,6))
sns.countplot(x='Subject', data = df)
```
